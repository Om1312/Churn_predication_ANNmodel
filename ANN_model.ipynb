Customer Churn Prediction using ANN (Artificial Neural Network)

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df = pd.read_csv(r'C:\all projects\data analysis project\deep_learning\dl_basics\churn predict\Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
input_data =  df.iloc[:,:-1]
output_data = df.iloc[:,-1]
```

```python
ss = StandardScaler()
input_data = pd.DataFrame(ss.fit_transform(input_data),columns = input_data.columns)
```

```python
input_data
```

```python
x_train,x_test,y_train,y_test = train_test_split(input_data,output_data,test_size = 0.2,random_state = 42)
```

```python
x_train.shape
```

```python
x_test
```

```python
import tensorflow
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann = Sequential()
```

```python
ann.add(Dense(6,input_dim = 8,activation = 'relu'))
ann.add(Dense(4,activation = 'relu'))
ann.add(Dense(2,activation = 'relu'))
ann.add(Dense(1,activation = 'sigmoid'))
```

```python
ann.compile(optimizer='adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size = 100,epochs = 50)
```

```python
prd1 = ann.predict(x_train)
prd_data1 = []
for i in prd1:
    if  i[0] > 0.5 :
        prd_data1.append(1)
    else :
        prd_data1.append(0)
 
```

```python
prd = ann.predict(x_test)
prd_data = []
for i in prd:
    if  i[0] > 0.5 :
        prd_data.append(1)
    else :
        prd_data.append(0)
 
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prd_data) *100
```

```python
accuracy_score(y_train,prd_data1) *100
```

```python
import numpy as np

# 1️⃣ New customer input (raw values in correct feature order)
# IMPORTANT: These must be in the SAME order as training features
input_raw = [[619,	42,	2,	0.00,	1,	1,	1,	101348.88 ]]

# 2️⃣ Convert to array
input_raw = np.array(input_raw)

# 3️⃣ Scale input exactly like training data
input_scaled = ss.transform(input_raw)

# 4️⃣ Predict probability of churn
prediction = ann.predict(input_scaled)

# 5️⃣ Convert probability → class
churn_class = (prediction > 0.5).astype(int)

# 6️⃣ Print results
print("Churn Probability:", prediction[0][0])
print("Churn Class (1 = Churn, 0 = Not Churn):", churn_class[0][0])
```
